### Data Preparation  

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
pd.set_option('display.max_columns', 29)

df = pd.read_excel('Data.xlsx')

display(df.head())
display(df.describe().T)
```

```python
df.dtypes
```

```python
print(f'shape:{df.shape}')

display(df['ID'].nunique())

display(np.any(df.isnull()))
```

```python
display(df.isna().sum())
```

```python
# Finding the rows with missing values 
i_missing = np.where(df.isna())[0]
df_miss = df.iloc[i_missing, :]
print(df_miss.shape)
display(df_miss)
```

```python
#Addign Age column 
df['Age'] =  2022 - df['Year_Birth'] 


#Removing the Year_Birth
df.drop('Year_Birth', axis = 1, inplace = True)
display(df.head())
```

```python
# We drop the dt_customer as it wont help us anyways. 

df.drop('Dt_Customer', axis = 1, inplace = True)

display(df.head())
display(df.shape)
```

```python
display(df['Education'].unique())

display(df['Marital_Status'].unique())
```

```python
## Reducing Education to 3 values only 

df["Education"]=df["Education"].replace({"Basic":"Undergraduate","2n Cycle":"Undergraduate", 
                                         "Graduation":"Graduate", "Master":"Postgraduate", "PhD":"Postgraduate"})

## Creating new column Married by Reducing Marital Status to married or unmarried creating a new column.
df["Married"] = df["Marital_Status"].apply(lambda x: 1 if (x == "Married" or x == "Together") else 0)

df = df.drop('Marital_Status', axis = 1)
```

```python
##combining teenhome and kidhome and new col: Children

df["Children"] = df["Kidhome"] + df['Teenhome']

df = df.drop('Kidhome', axis = 1)
df = df.drop('Teenhome', axis = 1)
df
```

```python
#df['Response'].unique()
df
```

```python
## Creating a column to total  'Mnt' columns together. 

df['time_spent'] = df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntSweetProducts'] + df['MntFishProducts'] + df['MntGoldProds'] 

## Creating a column to collect all purchases together 

df['total_purchase'] = df['NumDealsPurchases'] + df['NumWebPurchases'] + df['NumCatalogPurchases'] + df['NumStorePurchases'] 

## How many campaings accepted 

df['total_campaigns'] = df['AcceptedCmp3']  + df['AcceptedCmp2'] + df['AcceptedCmp1'] + df['AcceptedCmp5'] + df['AcceptedCmp4'] 


df
```

```python
## Complain columns
df['Complain'].value_counts()

## we see only 21 people had complains so this feature is highly skewed and wont be of use 
```

```python
df.drop(['Complain', 'AcceptedCmp3','AcceptedCmp2', 'AcceptedCmp1', 'AcceptedCmp5', 'AcceptedCmp4', 'Response', 'Z_CostContact', 'Z_Revenue', 'Recency', 'ID'], axis = 1, inplace = True)

df
```

```python
display(df.head())
display(df.shape)
```

```python
#encoding the education 

df['Education'] = df["Education"].replace({"Undergraduate":0, "Graduate":1, "Postgraduate":2})
```

```python
df['Education'].value_counts()
```

```python
# we use Knearest neighbour for imputation
from sklearn.impute import KNNImputer

imp = KNNImputer(n_neighbors=5)

df_knn = pd.DataFrame(imp.fit_transform(df))


display(df_knn)


np.any(df_knn.isna())

#to see imputation  - all income values have a valur now
print(f'KNN Imputed rows')
display(df_knn.iloc[i_missing, :])

```

```python
# Getting the column names back. 
ls = []
for i in df.columns:
    ls.append(i)
col = {}

#len(ls)
for i in range (0,len(ls)):
        col[i] = ls[i]
        
#print(col)

df_knn = df_knn.rename(columns=col)
#df_knn.head()
df = df_knn
display(df.head())
display(df.shape)
display(np.any(df.isnull()))
```
