```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
pd.set_option('display.max_columns', 29)

df = pd.read_excel('data.xlsx')

display(df.head())
display(df.describe().T)
```

```python
print(f'shape:{df.shape}')

display(df['ID'].nunique())

display(np.any(df.isnull()))
```

```python
# Missing values per column

display(df.isna().sum())
```

```python
# Finding the rows with missing values 
i_missing = np.where(df.isna())[0]
df_miss = df.iloc[i_missing, :]
print(df_miss.shape)
display(df_miss)
```

```python
# Lets see the distrubution of income

df['Income'].hist(bins= 150)
```

```python
#Addign Age column 
df['Age'] =  2022 - df['Year_Birth'] 


#Removing the Year_Birth
df.drop('Year_Birth', axis = 1, inplace = True)
display(df.head())
```

```python
# We drop the dt_customer as it wont help us anyways. 

df.drop('Dt_Customer', axis = 1, inplace = True)

display(df.head())
display(df.shape)
```

```python
display(df['Education'].unique())

display(df['Marital_Status'].unique())
```

```python
# encoding the education 

labels = df['Education'].astype('category').cat.categories.tolist()
edu = {'Education' : {k: v for k,v in zip(labels,list(range(1,len(labels)+1)))}}

print(edu)
```

```python
df.replace(edu, inplace=True)
display(df['Education'].unique())
```

```python
# encoding the Marital_Status

df = df.replace(['Single',
            'Together', 'Married', 'Divorced', 'Widow', 'Alone',
       'Absurd', 'YOLO'], [0, 1, 1, 0, 0, 0, 0, 0])

#print(ms)


#display(df['Marital_Status'].unique())
```

```python
display(df.head())
display(df.shape)
```

```python
#dropping the ID column
df = df.iloc[:,1:]
display(df.head())
display(df.shape)
```

```python
# we use Knearest neighbour for imputation
from sklearn.impute import KNNImputer

imp = KNNImputer(n_neighbors=5)

df_knn = pd.DataFrame(imp.fit_transform(df))


display(df_knn)


np.any(df_knn.isna())

#to see imputation  - all income values have a valur now
print(f'KNN Imputed rows')
display(df_knn.iloc[i_missing, :])

```

```python
# Getting the column names back. 
ls = []
for i in df.columns:
    ls.append(i)
col = {}

#len(ls)
for i in range (0,len(ls)):
        col[i] = ls[i]
        
#print(col)

df_knn = df_knn.rename(columns=col)
#df_knn.head()
df = df_knn
display(df.head())
display(df.shape)
```

```python
wine = df.iloc[:,6:7]
fruit = df.iloc[:,7:8]
meat = df.iloc[:,8:9]
fish = df.iloc[:,9:10]
sweet = df.iloc[:,10:11]
gold = df.iloc[:,11:12]

productInterests = df.iloc[:,6:12]

deal = df.iloc[:,12:13]
web = df.iloc[:,13:14]
catalog = df.iloc[:,14:15]
store = df.iloc[:,15:16]
webVisits = df.iloc[:,16:17]

productAvenues = df.iloc[:,12:17]

edu = df.iloc[:,0:1]
relationship = df.iloc[:,1:2]
income = df.iloc[:,2:3]
kids = df.iloc[:,3:4]
teens = df.iloc[:,4:5]


demographics = df.iloc[:,0:6]
age = df.iloc[:,-1:]
demographics.insert(0, 'Age', age)
```

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn import tree
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import Lars
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.linear_model import SGDClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

# Wine

```python
X_train, X_test, t_train, t_test =  train_test_split(demographics, wine, test_size=0.2, random_state=10)


#models
reg = linear_model.LinearRegression()
lasso = linear_model.Lasso()
ridge = linear_model.Ridge()
Enet = linear_model.ElasticNet()
sgd = linear_model.SGDClassifier()
lars = linear_model.Lars()
lassoLars = linear_model.LassoLars()
bRidge = linear_model.BayesianRidge()
logit = linear_model.LogisticRegression()
theil = linear_model.TheilSenRegressor()
huber = linear_model.HuberRegressor()

#training
reg.fit(X_train, t_train)
lasso.fit(X_train, t_train)
ridge.fit(X_train, t_train)
Enet.fit(X_train, t_train)
sgd.fit(X_train, t_train)
lars.fit(X_train, t_train)
lassoLars.fit(X_train, t_train)
bRidge.fit(X_train, t_train)
logit.fit(X_train, t_train)
theil.fit(X_train, t_train)
huber.fit(X_train, t_train)

#predictions
reg_y = reg.predict(X_test)
lasso_y = lasso.predict(X_test)
ridge_y = ridge.predict(X_test)
Enet_y = Enet.predict(X_test)
sgd_y = sgd.predict(X_test)

#scoring
print('Linear Train Accuracy : %.3f'%reg.score(X_train, t_train))
print('Linear Test Accuracy : %.3f'%reg.score(X_test, t_test))
print('Lasso Train Accuracy : %.3f'%lasso.score(X_train, t_train))
print('Lasso Test Accuracy : %.3f'%lasso.score(X_test, t_test))
print('Ridge Train Accuracy : %.3f'%ridge.score(X_train, t_train))
print('Ridge Test Accuracy : %.3f'%ridge.score(X_test, t_test))
print('Elastic Net Train Accuracy : %.3f'%Enet.score(X_train, t_train))
print('Elastic Net Test Accuracy : %.3f'%Enet.score(X_test, t_test))
print('SGD Train Accuracy : %.3f'%sgd.score(X_train, t_train))
print('SGD Test Accuracy : %.3f'%sgd.score(X_test, t_test))
print('Lars Train Accuracy : %.3f'%lars.score(X_train, t_train))
print('Lars Test Accuracy : %.3f'%lars.score(X_test, t_test))
print('Lasso Lars Train Accuracy : %.3f'%lassoLars.score(X_train, t_train))
print('Lasso Lars Test Accuracy : %.3f'%lassoLars.score(X_test, t_test))
print('Bayesian Ridge Train Accuracy : %.3f'%bRidge.score(X_train, t_train))
print('Bayesian Ridge Test Accuracy : %.3f'%bRidge.score(X_test, t_test))
print('Logistic Regression Train Accuracy : %.3f'%logit.score(X_train, t_train))
print('Logistic Regression Test Accuracy : %.3f'%logit.score(X_test, t_test))
print('Thiel Sen Regression Train Accuracy : %.3f'%theil.score(X_train, t_train))
print('Theil Sen Regression Test Accuracy : %.3f'%theil.score(X_test, t_test))
print('Huber Regression Train Accuracy : %.3f'%huber.score(X_train, t_train))
print('Huber Regression Test Accuracy : %.3f'%huber.score(X_test, t_test))
```
